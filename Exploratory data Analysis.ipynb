```python
continent_name = 'Africa'

continent_data = df[df['continent_name'] == continent_name]

# Calculate the sum of each salary metric for the continent
total_highest_salary = continent_data['highest_salary'].sum()
total_lowest_salary = continent_data['lowest_salary'].sum()
total_median_salary = continent_data['median_salary'].sum()
total_average_salary = continent_data['average_salary'].sum()

labels = ['Highest Salary', 'Lowest Salary', 'Median Salary', 'Average Salary']
sizes = [total_highest_salary, total_lowest_salary, total_median_salary, total_average_salary]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.1188, 0.155, 0.1123)  

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title(f'Distribution of Salary Metrics in {continent_name}')
plt.axis('equal') 

plt.show()
```
